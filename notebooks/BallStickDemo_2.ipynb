# Ball Stick Model Example
## Magnetic Resonance Imaging (MRI)
Magnetic Resonance Imaging (MRI) is a non-invasive imaging technique that uses strong magnetic fields and radio waves to generate detailed images of the human body.
Unlike CT or X-ray, MRI does not involve ionizing radiation, making it safer for repeated clinical use.

## Diffusion MRI
Diffusion MRI (dMRI) is a specialized form of MRI that measures the random motion of water molecules within tissue.
Since water diffusion is affected by the microscopic structure of tissues, dMRI provides unique information about tissue organization that is not visible in conventional MRI.

- **Diffusion-Weighted Imaging (DWI)**: Sensitive to the degree of water diffusion.
- **Diffusion Tensor Imaging (DTI)**: Models diffusion as a tensor, allowing estimation of anisotropy and fiber orientation.

This makes dMRI particularly valuable in studying **white matter in the brain**, where axonal bundles constrain water motion along preferred directions.

## What is Ball-Stick Model
One of the simplest yet powerful models for diffusion MRI is the **Ball–Stick model**:
- **Stick**: Represents highly anisotropic diffusion along a single direction (e.g., inside axon bundles).
- **Ball**: Represents isotropic diffusion in extracellular space.

By fitting the Ball–Stick model to diffusion MRI data, we can estimate:
- The **orientation** of dominant fibers.
- The **volume fraction** between anisotropic and isotropic components.

## How does this model works

Traditionally, fitting the Ball–Stick model requires solving a nonlinear optimization problem to minimize the difference between measured diffusion signals and model predictions. This process can be computationally expensive and sensitive to initialization.

With **self-supervised learning**, we treat the model parameters (e.g., fiber orientation, volume fraction, diffusivities) as *trainable variables* in a neural network. Instead of using labeled data, the network learns directly from the raw diffusion signals by minimizing a reconstruction loss:

$ \mathcal{L} = \| S_\text{measured} - S_\text{predicted}(\theta) \|^2 $

where:
- $ S_\text{measured} $ = the observed diffusion-weighted signal
- $ S_\text{predicted}(\theta) $ = the Ball–Stick model output given current parameters $ \theta $

The optimization loop is similar to neural network training:
1. Initialize parameters randomly.
2. Forward pass: compute predicted signal from the Ball–Stick model.
3. Compute loss function (mean squared error).
4. Backpropagate gradients and update parameters using gradient descent.
5. Iterate until convergence.

This way, the Ball–Stick parameters are learned **without explicit supervision**, only by matching the predicted signal to the measured diffusion MRI data.


## Learning parameters
#### 1. `random_seed`
- **Definition**: A fixed number used to initialize random number generators.
- **Purpose**: Ensures reproducibility of experiments. With the same seed, data shuffling, weight initialization, and other stochastic processes will produce the same results across runs.

---

#### 2. `learning_rate`
- **Definition**: The step size used in gradient descent optimization.
- **Purpose**: Controls how much the model’s parameters are updated during training.
- **Notes**:
  - Too high → unstable training or divergence.
  - Too low → very slow convergence.

---

#### 3. `activation`
- **Definition**: The nonlinear function applied to neuron outputs.
- **Purpose**: Introduces nonlinearity, allowing the model to learn complex representations.
- **Common choices**: `ReLU`, `Sigmoid`, `Tanh`, `LeakyReLU`.
- **Notes**: The choice of activation affects training dynamics and the expressiveness of the model.

---

#### 4. `iteration`
- **Definition**: Number of passes over the training dataset (sometimes also called "epochs").
- **Purpose**: Determines how long the model is trained.
- **Notes**:
  - Too few iterations → underfitting.
  - Too many iterations → risk of overfitting (especially without regularization).

---

#### 5. `dropout_fraction`
- **Definition**: The fraction (0–1) of neurons randomly "dropped" during training.
- **Purpose**: Acts as a regularization method to prevent overfitting.
- **Example**: A dropout fraction of 0.5 means half of the neurons are temporarily disabled in each training step.
- **Notes**: Dropout is only applied during training, not during inference.
