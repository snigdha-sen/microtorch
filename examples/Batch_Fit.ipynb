- Note
    The file structure in selected folder should be:
    ```
        /selected
            |---/sample_1
            |      |--- *.nii.gz         // Image file
            |      |--- *_mask.nii.gz    // Mask file
            |      |--- grad.txt         // Gradient file (Optional if you have following files)
            |      |--- bvals            // b Values file
            |      |--- bvecs            // b Vectors file
            |
            |---/sample_2
                   |--- *.nii.gz         // Image file
                   |--- *_mask.nii.gz    // Mask file
                   |--- grad.txt         // Gradient file (Optional if you have following files)
                   |--- bvals            // b Values file
                   |--- bvecs            // b Vectors file
    ```
    Or:
    ```
        /selected
            |--- *.nii.gz         // Image file
            |--- *_mask.nii.gz    // Mask file
            |--- grad.txt         // Gradient file (Optional if you have following files)
            |--- bvals            // b Values file
            |--- bvecs            // b Vectors file
    ```

    Please make sure **Mask** file is named as `***_mask.nii.gz` , **b Values** and **b Vectors** files are named as `bvecs` and `bvals`.
