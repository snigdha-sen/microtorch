# fit.py
This notebook is a step-by-step interactive GUI of `fit.py`.

- Fit Steps

    `Select Model` ==> `Choose Files` ==> `Check Image` ==> `Set Training Parameters` ==> `Training` ==> `Check Results`

- Note

    1. If you chose a 3-Dimension image file, the `file_viewer` will pick a slice to both display and training.
    2. After training with selected dataset, you can select and load another dataset, and then match the new files with the trained net directly.
    2. This notebook **won't** save any parameter maps.

```python
import utils.utils_jupyter as utils
from IPython.display import display

files = {}
file_loader = utils.FileLoader(files)
file_viewer = utils.DataViewer()
net_trainer = utils.TrainingPanel()
result_viewer = utils.ResultVisualizer()

model_chooser = utils.ModelChooser(file_loader)

file_loader.register(file_viewer)
file_loader.register(net_trainer)
net_trainer.register(result_viewer)

display(model_chooser.ui)
display(file_loader.ui)
display(file_viewer.ui)
display(net_trainer.ui)
display(result_viewer.ui)
```
